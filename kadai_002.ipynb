```python
#課題：スクレイピング(kadai_02)

#Step1
from bs4 import BeautifulSoup
from urllib import request
import re
from janome.tokenizer import Tokenizer

url = "https://www.aozora.gr.jp/cards/000148/files/2371_13943.html"

response = request.urlopen(url)                        #文章の取得
soup = BeautifulSoup(response)
response.close()
#print(soup)
main_text = soup.find("div",  class_ ="main_text")         #本文の取得
#print(main_text)

tags_to_delete = main_text.find_all(["rp", "rt"])            #ルビ関連のタグ削除
for tag in tags_to_delete:
    tag.decompose()
print(main_text)
```

```python
#Step2：HTMLタグや不要な文字の削除

main_text = main_text.get_text()                          #本文の抽出
#print(main_text)

main_text = re.sub(r"[\u3000 『』　　―　\n \r]", "", main_text)     #テキストクリーニング
print(main_text)
```

```python
#Step3：ストップワードの除去

# 文章を品詞ごとに分割し、助詞を削除して名詞のリストを得る
def tokenize_text(text):
    tokenizer = Tokenizer()
    tokens = tokenizer.tokenize(text)
    return [token.surface for token in tokens if '助詞' not in token.part_of_speech]

result_tokens = tokenize_text(main_text)
#print(result_tokens)

#SlothLibのテキストページからストップワードの辞書を取得

url = "http://svn.sourceforge.jp/svnroot/slothlib/CSharp/Version1/SlothLib/NLP/Filter/StopWord/word/Japanese.txt"

response2 = request.urlopen(url)
soup2 = BeautifulSoup(response2)
response2.close()
stopwords_text = soup2.text

stopwords_list = stopwords_text.split("\r\n")
stopwords_list = [word for word in stopwords_list if word]
#print(stopwords_list)

#辞書ベースのストップワードによる除去
result_text_list = list()
for split_text in result_tokens:
  if split_text not in stopwords_list:
    result_text_list.append(split_text)

print(result_text_list)
```
